```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Printing the whole DataFrame
df = pd.read_csv('./data/day.csv', delimiter=',')
df
```

```python
df.describe() # Overall data description
```

```python
# Elements summing using DataFrame.loc
temperature = df.loc[1:5, 'temp']
humidity = df.loc[1:5, 'hum']

temperature
humidity

sum = temperature + humidity
sum
```

```python
# Renaming the 'mnth' column to 'month'
df.rename(columns={'mnth':'month'}, inplace=True)
df.head(1)
```

```python
# Average temperature in Semptember
sep_temp_avg = df.loc[df['month'] == 9, 'temp'].mean()
sep_temp_avg
```

```python
# Get rented bikes count grouped by humidity average value
df['hum'].groupby(df['cnt']).mean()
```

```python
# Data visualization

temperature = df['temp']
rented_cnt = df['cnt']

plt.title("Example scattered graph", fontsize=15, color="green")
plt.ylabel("Temperature")
plt.xlabel("Bikes rented count")

plt.scatter(rented_cnt, temperature, color='black')
```
