```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.preprocessing import minmax_scale
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('./data/glass.csv', delimiter=',')

var_names = list(data.columns)
labels = data.to_numpy('int')[:, -1]
features = data.to_numpy('float')[:, : -1]

normalize(features, norm="l2")
minmax_scale(features, axis=1)
```

```python
pca = PCA(n_components=2)
pca_data = pca.fit(features)
pca_data = pca.transform(features)
```

```python
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

```python
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=labels, cmap='hsv')
```
